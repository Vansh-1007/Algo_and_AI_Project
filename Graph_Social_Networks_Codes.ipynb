The Karate dataset is a well-known social network dataset that represents the friendships among members of a karate club at a US university in the 1970s. It was originally studied by Wayne W. Zachary, who analyzed the social interactions among the members to understand the dynamics of the group
Nodes:

Edges:

Edges represent friendships between the members. If two members are friends, there is an edge connecting them

Graph Representation:

The dataset can be represented as an undirected graph, where nodes represent members and edges represent the friendships.

K-Clans

```python
from itertools import combinations

# Function to find k-clans
def find_k_clans(G, k):
    clans = []
    for node in G.nodes:
        neighbors = list(G.neighbors(node))
        for clan in combinations(neighbors, k):
            if all(G.has_edge(clan[i], clan[j]) for i in range(len(clan)) for j in range(i + 1, len(clan))):
                clans.append(clan)
    return clans

k_clans = find_k_clans(G, k_value)
print(f"k-Clans (k={k_value}): {k_clans}")
```

K-clubs

```python
# Function to find k-clubs
def find_k_clubs(G, k):
    clubs = []
    for node in G.nodes:
        neighbors = list(G.neighbors(node))
        club = [node]
        for neighbor in neighbors:
            if len(nx.shortest_path(G, source=node, target=neighbor)) <= k:
                club.append(neighbor)
        clubs.append(club)
    return [list(set(club)) for club in clubs] 

k_clubs = find_k_clubs(G, k_value)
print(f"k-Clubs (k={k_value}): {k_clubs}")
```

K-plexes

```python
# Function to find k-plexes
def find_k_plexes(G, k):
    plexes = []
    for node in G.nodes:
        neighbors = list(G.neighbors(node))
        for plex in combinations(neighbors, k):
            if all(len(set(plex).intersection(set(G.neighbors(neighbor)))) >= (k - 1) for neighbor in plex):
                plexes.append(plex)
    return plexes 
    
k_plexes = find_k_plexes(G, k_value)
print(f"k-Plexes (k={k_value}): {k_plexes}")
```

SIS (Susceptible-Infected-Susceptible)  model

```python

def sis_model(graph, beta, gamma, initial_infected, time_steps):
    # Initialize the states
    states = {node: 'S' for node in graph.nodes()}
    # Randomly select initial infected individuals
    infected_nodes = np.random.choice(graph.nodes(), initial_infected, replace=False)
    for node in infected_nodes:
        states[node] = 'I'

    # Store the number of infected individuals over time
    infected_counts = []

    for t in range(time_steps):
        new_infected = []
        for node in graph.nodes():
            if states[node] == 'I':
                # Each infected node has a chance to infect its neighbors
                for neighbor in graph.neighbors(node):
                    if states[neighbor] == 'S' and np.random.rand() < beta:
                        new_infected.append(neighbor)
                # Each infected node recovers
                if np.random.rand() < gamma:
                    states[node] = 'S'
        
        # Update the states
        for node in new_infected:
            states[node] = 'I'
        
        # Count the number of infected nodes
        infected_counts.append(sum(1 for state in states.values() if state == 'I'))

    return infected_counts

# Load the Karate Club dataset
karate_graph = nx.karate_club_graph()

# Parameters for the SIS model
beta = 0.3  # Infection probability
gamma = 0.1  # Recovery probability
initial_infected = 1  # Initial infected individuals
time_steps = 50  # Number of time steps to simulate

# Run the SIS model
infected_counts_sis = sis_model(karate_graph, beta, gamma, initial_infected, time_steps)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(infected_counts_sis, label='Infected Individuals', color='red')
plt.title('SIS Model Simulation on Karate Club Graph')
plt.xlabel('Time Steps')
plt.ylabel('Number of Infected Individuals')
plt.legend()
plt.grid()
plt.show()
```

SEIS (Susceptible-Exposed-Infected-Susceptible) model

```python
def seis_model(graph, beta, sigma, gamma, initial_infected, time_steps):
    # Initialize the states
    states = {node: 'S' for node in graph.nodes()}
    infected_nodes = np.random.choice(graph.nodes(), initial_infected, replace=False)
    for node in infected_nodes:
        states[node] = 'I'

    infected_counts = []

    for t in range(time_steps):
        new_infected = []
        new_exposed = []
        
        for node in graph.nodes():
            if states[node] == 'I':
                # Each infected node has a chance to infect its neighbors
                for neighbor in graph.neighbors(node):
                    if states[neighbor] == 'S' and np.random.rand() < beta:
                        new_exposed.append(neighbor)
                # Recover infected individuals
                if np.random.rand() < gamma:
                    states[node] = 'S'
        
        # Update states
        for node in new_exposed:
            states[node] = 'E'
        
        # Transition from exposed to infected
        for node in graph.nodes():
            if states[node] == 'E' and np.random.rand() < sigma:
                states[node] = 'I'

        # Count the number of infected individuals
        infected_counts.append(sum(1 for state in states.values() if state == 'I'))

    return infected_counts

# Parameters for the SEIS model
beta = 0.3   # Infection probability
sigma = 0.2  # Exposure probability
gamma = 0.1   # Recovery probability
initial_infected = 1  # Initial infected individuals
time_steps = 50  # Number of time steps to simulate

# Run the SEIS model
infected_counts_seis = seis_model(karate_graph, beta, sigma, gamma, initial_infected, time_steps)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(infected_counts_seis, label='Infected Individuals', color='blue')
plt.title('SEIS Model Simulation on Karate Club Graph')
plt.xlabel('Time Steps')
plt.ylabel('Number of Infected Individuals')
plt.legend()
plt.grid()
plt.show()
```

Entropy 

```python
def calculate_entropy(graph):
    # Calculate the degree of each node
    degrees = [degree for node, degree in graph.degree()]
    
    # Calculate the degree distribution
    degree_counts = np.bincount(degrees)
    degree_probs = degree_counts / np.sum(degree_counts)
    
    # Calculate entropy
    entropy = -np.sum(degree_probs[degree_probs > 0] * np.log(degree_probs[degree_probs > 0]))
    
    return entropy

# Load the Karate Club dataset
G = nx.karate_club_graph()

# Calculate the entropy of the Karate Club graph
entropy_karate = calculate_entropy(G)

print(f"Entropy of the Karate Club Graph: {entropy_karate:.4f}")
```

Sim Rank 

```python

def simrank(graph, C=0.8, max_iter=100, tol=1e-5):
    # Initialize SimRank matrix
    sim_matrix = np.zeros((len(graph), len(graph)))

    # Create a mapping from nodes to indices
    node_to_index = {node: i for i, node in enumerate(graph.nodes())}
    
    # Iteratively update the SimRank values
    for _ in range(max_iter):
        prev_sim_matrix = sim_matrix.copy()
        
        for a in graph.nodes():
            for b in graph.nodes():
                if a != b:
                    # Get neighbors
                    neighbors_a = list(graph.neighbors(a))
                    neighbors_b = list(graph.neighbors(b))
                    
                    # Calculate SimRank
                    if len(neighbors_a) > 0 and len(neighbors_b) > 0:
                        sum_sim = 0.0
                        for x in neighbors_a:
                            for y in neighbors_b:
                                sum_sim += prev_sim_matrix[node_to_index[x], node_to_index[y]]
                        sim_matrix[node_to_index[a], node_to_index[b]] = (C / (len(neighbors_a) * len(neighbors_b))) * sum_sim
        
        # Check for convergence
        if np.max(np.abs(sim_matrix - prev_sim_matrix)) < tol:
            break
            
    return sim_matrix

# Load the Karate Club dataset
karate_graph = nx.karate_club_graph()

# Calculate SimRank
simrank_matrix = simrank(karate_graph)

# Example: Get SimRank between nodes 0 and 1
node_a = 5
node_b = 8
simrank_value = simrank_matrix[node_a, node_b]
print(f"SimRank between nodes {node_a} and {node_b}: {simrank_value:.4f}")
```

Dunbar number

```python


def dunbar_number(graph):
    # Calculate the average degree of the nodes
    degrees = [degree for node, degree in graph.degree()]
    average_degree = sum(degrees) / len(degrees)
    
    # Calculate Dunbar's number based on average degree
    # The Dunbar number is often approximated to be around 3 times the average degree.
    dunbar_num = average_degree * 3
    
    return round(dunbar_num)

# Load the Karate Club dataset
karate_graph = nx.karate_club_graph()

# Calculate the Dunbar number for the Karate Club graph
dunbar_num = dunbar_number(karate_graph)

print(f"Dunbar Number for the Karate Club Graph: {dunbar_num}")

# Visualize the graph
plt.figure(figsize=(8, 6))
nx.draw(karate_graph, with_labels=True, node_color='skyblue', node_size=500, font_size=10)
plt.title('Karate Club Graph')
plt.show()
```

Neighbourhood Overlap



```python
import pandas as pd

def neighborhood_overlap(graph):
    overlap_matrix = pd.DataFrame(index=graph.nodes(), columns=graph.nodes()).fillna(0)

    for a in graph.nodes():
        for b in graph.nodes():
            if a != b:
                neighbors_a = set(graph.neighbors(a))
                neighbors_b = set(graph.neighbors(b))
                
                # Calculate overlap
                intersection_size = len(neighbors_a & neighbors_b)
                union_size = len(neighbors_a | neighbors_b)
                
                if union_size > 0:
                    overlap_matrix.loc[a, b] = intersection_size / union_size

    return overlap_matrix

# Load the Karate Club dataset
karate_graph = nx.karate_club_graph()

# Calculate neighborhood overlap
overlap_matrix = neighborhood_overlap(karate_graph)

# Display the overlap between two specific nodes, for example, nodes 0 and 1
node_a = 0
node_b = 1
overlap_value = overlap_matrix.loc[node_a, node_b]
print(f"Neighborhood overlap between nodes {node_a} and {node_b}: {overlap_value:.4f}")

# Display the entire overlap matrix (optional)
print("\nNeighborhood Overlap Matrix:")
print(overlap_matrix)
```
