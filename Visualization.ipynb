```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the karate club dataset
G = nx.karate_club_graph()

# Calculate shortest path lengths from node 0
shortest_path_from_node_0 = nx.shortest_path_length(G, source=0)

# Define colors based on the shortest path lengths
color_map = []
for node in G.nodes():
    if shortest_path_from_node_0[node] == 0:
        color_map.append('lightcoral')  # Node 0 itself
    elif shortest_path_from_node_0[node] == 1:
        color_map.append('lightblue')   # Distance 1
    elif shortest_path_from_node_0[node] == 2:
        color_map.append('lightgreen')  # Distance 2
    elif shortest_path_from_node_0[node] == 3:
        color_map.append('lightyellow') # Distance 3

# Plotting the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Use spring layout for better visualization

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_color=color_map, node_size=500, edgecolors="black")
nx.draw_networkx_edges(G, pos, alpha=0.3)
nx.draw_networkx_labels(G, pos, font_size=10, font_color="black")

# Add a legend
legend_labels = {
    'lightcoral': 'Node 0',
    'lightblue': 'Distance 1',
    'lightgreen': 'Distance 2',
    'lightyellow': 'Distance 3'
}
for color, label in legend_labels.items():
    plt.plot([], [], color=color, marker='o', linestyle='', markersize=10, label=label)
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, loc='upper right')

# Title and show the plot
plt.title('Shortest Path Lengths from Node 0 in Karate Club Network')
plt.axis('off')  # Turn off the axis
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the karate club graph
G = nx.karate_club_graph()

def bfs(graph, start_node):
    visited = set()
    queue = [start_node]
    bfs_order = []

    while queue:
        node = queue.pop(0)  # Dequeue the first node
        if node not in visited:
            visited.add(node)  # Mark the node as visited
            bfs_order.append(node)  # Keep track of the BFS order
            # Enqueue all unvisited neighbors
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    queue.append(neighbor)

    return bfs_order

# Run BFS starting from node 0
start_node = 0
bfs_order = bfs(G, start_node)

# Create a mapping of node to its BFS order
order_mapping = {node: i for i, node in enumerate(bfs_order)}

# Visualize the graph
pos = nx.spring_layout(G)  # positions for all nodes

# Draw the entire graph
nx.draw(G, pos, with_labels=True, node_color='lightgray', node_size=600, font_size=10)

# Highlight the BFS traversal
bfs_edges = [(bfs_order[i], bfs_order[i + 1]) for i in range(len(bfs_order) - 1)]
nx.draw_networkx_edges(G, pos, edgelist=bfs_edges, edge_color='red', width=2)

# Annotate nodes with their order of visitation
for node, order in order_mapping.items():
    plt.text(pos[node][0], pos[node][1], str(order), fontsize=12, ha='center', color='black')

# Display the BFS order
print("BFS Order:", bfs_order)

# Show the plot
plt.title("BFS Traversal of the Karate Club Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the karate club graph
G = nx.karate_club_graph()

def bfs(graph, start_node):
    visited = set()
    queue = [start_node]
    bfs_order = []

    while queue:
        node = queue.pop(0)  # Dequeue the first node
        if node not in visited:
            visited.add(node)  # Mark the node as visited
            bfs_order.append(node)  # Keep track of the BFS order
            # Enqueue all unvisited neighbors
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    queue.append(neighbor)

    return bfs_order

# Run BFS starting from node 0
start_node = 0
bfs_order = bfs(G, start_node)

# Create a mapping of node to its BFS order
order_mapping = {node: i for i, node in enumerate(bfs_order)}

# Set the figure size for better visibility
plt.figure(figsize=(15, 15))  # Adjust the figure size

# Use spring layout with increased distance between nodes
pos = nx.spring_layout(G, k=0.5)  # You can adjust the k parameter

# Set the size for the nodes
node_size = 1000  # Adjust node size as needed

# Draw the entire graph
nx.draw(G, pos, with_labels=True, node_color='lightgray', node_size=node_size, font_size=10)

# Highlight the BFS traversal
bfs_edges = [(bfs_order[i], bfs_order[i + 1]) for i in range(len(bfs_order) - 1)]
nx.draw_networkx_edges(G, pos, edgelist=bfs_edges, edge_color='red', width=2)

# Annotate nodes with their order of visitation, with offsets to prevent overlap
for node, order in order_mapping.items():
    plt.text(pos[node][0], pos[node][1] + 0.1, str(node), fontsize=12, ha='center', color='black')  # Node number
    plt.text(pos[node][0], pos[node][1] - 0.1, str(order), fontsize=10, ha='center', color='blue')  # Order number

# Display the BFS order
print("BFS Order:", bfs_order)

# Show the plot
plt.title("BFS Traversal of the Karate Club Graph")
plt.show()
```

```python
import networkx as nx
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the karate club graph
G = nx.karate_club_graph()

# Calculate adjacency matrix
adjacency_matrix = nx.to_numpy_array(G)

# Function to calculate SimRank
def simrank(adjacency_matrix, c=0.8, max_iter=100, tol=1e-5):
    n = adjacency_matrix.shape[0]
    sim = np.zeros((n, n))
    prev_sim = np.copy(sim)

    for iteration in range(max_iter):
        for i in range(n):
            for j in range(n):
                if i != j:
                    # Calculate similarity based on neighbors
                    sim[i, j] = c * np.sum(
                        [prev_sim[n1, n2] for n1 in np.where(adjacency_matrix[i] > 0)[0]
                         for n2 in np.where(adjacency_matrix[j] > 0)[0]]
                    ) / (np.sum(adjacency_matrix[i]) * np.sum(adjacency_matrix[j]) + 1e-10)

        # Check for convergence
        if np.linalg.norm(sim - prev_sim) < tol:
            break
        prev_sim = np.copy(sim)

    return sim

# Calculate SimRank
similarity_matrix = simrank(adjacency_matrix)

# Print the similarity matrix to debug
print("SimRank Similarity Matrix:")
print(similarity_matrix)

# Normalize similarity scores for better visualization

# Normalize similarity scores for better visualization, with safe min/max
min_sim = np.min(similarity_matrix)
max_sim = np.max(similarity_matrix)

# Only normalize if max_sim is greater than min_sim
if max_sim > min_sim:
    normalized_similarity_matrix = (similarity_matrix - min_sim) / (max_sim - min_sim)
else:
    normalized_similarity_matrix = similarity_matrix  # Keep original if no range

# Create a heatmap as before


# Create a heatmap for SimRank similarity scores
plt.figure(figsize=(10, 8))
sns.heatmap(normalized_similarity_matrix, cmap='plasma', annot=True, fmt=".2f", square=True,
            cbar_kws={"shrink": .8}, xticklabels=G.nodes(), yticklabels=G.nodes(),
            annot_kws={"size": 10}, linewidths=0.5, linecolor='black')

# Add titles and labels
plt.title('SimRank Similarity Heatmap for the Karate Club Graph', fontsize=18)
plt.xlabel('Node', fontsize=14)
plt.ylabel('Node', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yticks(rotation=0, ha='right')    # Keep y-axis labels horizontal

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
